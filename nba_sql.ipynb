```python
import numpy as np
import pandas as pd
import psycopg2 
from sqlalchemy import create_engine 
import os 
POSTGRES_PASSWORD = os.getenv("POSTGRES_PASSWORD")
```

```python
engine = create_engine("postgresql+psycopg2://{user}:{pw}@{service}:{port}/{db}".format(
    user = 'postgres', pw = POSTGRES_PASSWORD, service= "postgres",port="5432", db = 'nba', 
))
```

```python
myquery = '''
SELECT 
game_date AS date, 
season AS nba_season, 
ot AS OT,
2*ot AS two_times_ot
FROM games 
WHERE season = 2022 AND ot>2
ORDER BY ot DESC
LIMIT 20
'''

pd.read_sql_query(myquery,con = engine)
```

from the player_game table, make a table with all players who scored more than 30 points in the game, Keep
only the columns with the game and player id, points, and three point shots made and attempted, sort by points in the descending order and display just the top 15

```python
query = '''
SELECT 
player_id, pts, fg3, fg3a
FROM player_game
WHERE pts>30
ORDER BY pts DESC
LIMIT 15
'''
pd.read_sql_query(query, con = engine)
```

```python
query1 = '''

SELECT g.game_date, p.player, pg.pts, pg.fg3, pg.fg3a
FROM player_game pg
INNER JOIN players p
    ON pg.player_id = p.player_id
INNER JOIN games g
    ON pg.game_id = g.game_id
WHERE pg.pts>30 
ORDER BY pg.pts DESC
LIMIT 15
'''
```

```python
pd.read_sql_query(query1, con = engine)
```

in the team_game table, who won each game. that will require you to join the team game to itself. keep the game id, the team and opponent abbreviation, the team score, and the other team score and rename to 'opponent score'

```python
query2 = '''
SELECT 
    tg.game_id, 
    tg.team_abbrev, 
    tg.opponent_abbrev, 
    tg.team_score, 
    tg1.team_score AS opponent_score,
    tg.team_score>tg1.team_score AS win
FROM team_game tg
INNER JOIN team_game tg1
    ON tg.game_id = tg1.game_id AND tg.opponent_abbrev = tg1.team_abbrev
'''
```

```python
pd.read_sql_query(query2, con = engine)
```
