## Bring the voteview data and filter to the 118th congress

```python
ideology = pd.read_csv("https://voteview.com/static/data/out/members/HS118_members.csv")
```

```python
ideology["chamber"].value_counts()
```

```python
ideology[["bioname",'chamber','party_code','district_code','state_abbrev','nominate_dim1']]\
.sort_values("nominate_dim1",ascending=True, ignore_index = True)


# ['bioname'].sort_values['nominate_dim1']
```

```python
replace_map = {100:'Democrat',200:'Republican',328:"independent"}

ideology['party_code'] = ideology['party_code'].replace(replace_map)
ideology["party_code"].value_counts()
```

```python
ideology.query("party_code == 'independent'")[['bioname','chamber','state_abbrev','nominate_dim1']]
```

```python
ideology.query("state_abbrev == 'VA' & chamber == 'House'")[['district_code','party_code','bioname']]
```

```python
votes = pd.read_csv("https://voteview.com/static/data/out/votes/HS118_votes.csv")
```

```python
# first do the merge in a way to checks for problems
ideology_votes = pd.merge(votes,ideology,on = 'icpsr', how = 'outer', validate = 'many_to_one',indicator = 'matched')
```
